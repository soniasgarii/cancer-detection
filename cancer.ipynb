```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib

# نمایش نمودارها داخل نوت‌بوک
%matplotlib widget

# خواندن دیتاست
data = pd.read_csv("cancer.csv")

# حذف ستون‌های غیرضروری
data = data.drop(["Unnamed: 32", "id"], axis=1)

# تبدیل برچسب‌ها
# M = 1 (سرطانی)
# B = 0 (خوش‌خیم)
data["diagnosis"] = [1 if value == "M" else 0 for value in data["diagnosis"]]

# جدا کردن خروجی (Label)
y = data["diagnosis"]

# جدا کردن ویژگی‌ها (Features)
X = data.drop(["diagnosis"], axis=1)

# نرمال‌سازی داده‌ها بین 0 و 1
X = (X - np.min(X)) / (np.max(X) - np.min(X))

from sklearn.model_selection import train_test_split

# تقسیم داده‌ها به آموزش و آزمون
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.15,
    random_state=42
)

from sklearn.linear_model import LogisticRegression

# ساخت مدل Logistic Regression
model = LogisticRegression()

# افزایش تعداد تکرار برای جلوگیری از خطای همگرایی
model.max_iter = 100000000

# آموزش مدل
model.fit(X_train, y_train)

# ذخیره مدل آموزش‌دیده
joblib.dump(model, "breast_cancer_model.pkl")

# پیش‌بینی روی داده‌های آزمون
prediction = model.predict(X_test)

# مقایسه پیش‌بینی با جواب واقعی
result = (prediction == y_test)

# تبدیل به DataFrame برای ریست ایندکس
result = result.reset_index(drop=True)

# شمارش تعداد پیش‌بینی‌های اشتباه
error_count = sum(result == False)

# نمایش تعداد خطاها
print("Wrong Predictions:", error_count)
print("Total Test Samples:", len(result))

# محاسبه دقت مدل
accuracy = model.score(X_test, y_test)

print("Accuracy:", accuracy)
```
