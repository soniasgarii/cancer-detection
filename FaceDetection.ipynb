```python
import cv2

# اتصال به دوربین
camera = cv2.VideoCapture(0)

# مدل تشخیص چهره
face_detector = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

# مدل تشخیص چشم
eye_detector = cv2.CascadeClassifier("haarcascade_eye.xml")

while True:

    # دریافت یک فریم از دوربین
    ret, frame = camera.read()

    # بررسی دریافت تصویر
    if not ret:
        print("Cannot read from camera")
        break

    # تبدیل تصویر به خاکستری
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # تشخیص چهره
    faces = face_detector.detectMultiScale(gray_frame)

    # کپی تصویر اصلی
    output = frame.copy()

    # اگر هیچ چهره‌ای پیدا نشد
    if len(faces) == 0:

        # نمایش پیام روی تصویر
        cv2.putText(
            output,
            "No Face Detected",
            (20,40),
            cv2.FONT_HERSHEY_SIMPLEX,
            1,
            (0,0,255),
            2
        )

    else:

        # مختصات گوشه بالا-چپ صورت
        face_x = faces[0][0]
        face_y = faces[0][1]

        # عرض و ارتفاع صورت
        face_width = faces[0][2]
        face_height = faces[0][3]

        # محاسبه گوشه پایین-راست مستطیل
        face_x2 = face_x + face_width
        face_y2 = face_y + face_height

        # جدا کردن قسمت صورت
        gray_face = gray_frame[face_y:face_y2, face_x:face_x2]

        # تشخیص چشم
        eyes = eye_detector.detectMultiScale(gray_face)

        # رسم مستطیل دور صورت
        cv2.rectangle(
            output,
            (face_x, face_y),
            (face_x2, face_y2),
            (255,0,0),
            2
        )

        # رسم مستطیل دور چشم‌ها
        for (eye_x, eye_y, eye_width, eye_height) in eyes:

            cv2.rectangle(
                output,
                (eye_x+face_x, eye_y+face_y),
                (eye_x+eye_width+face_x, eye_y+eye_height+face_y),
                (0,255,0),
                2
            )

    # نمایش خروجی
    cv2.imshow("Face Detection", output)

    # خروج با زدن q
    if cv2.waitKey(1) == ord('q'):
        break

camera.release()
cv2.destroyAllWindows()
```
